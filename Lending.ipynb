```python
# Importing Dependencies
import pandas as pd
import numpy as np
```

# Data Preproccessing

```python
# Import the 2019 Data (Training Data)
df_2019 = pd.read_csv('Resources/2019loans.csv')
df_2019.head()
```

```python
# Converting Categorical Data into numeric
df_2019_dummies = pd.get_dummies(df_2019)
df_2019_dummies.head()
```

```python
# Extracting the target Features [loan_status_low_risk , loan_status_high_risk]

target_features = df_2019['loan_status'].unique() # finding the unique categorical value to be converted into columns

X_train = df_2019_dummies.drop(['loan_status_low_risk' , 'loan_status_high_risk' , 'Unnamed: 0'] , axis=1)
y_train = df_2019_dummies[ 'loan_status_high_risk']
```

```python
# Reading in 2020 data
df_2020 = pd.read_csv('Resources/2020Q1loans.csv')
df_2020.head()

# Dropping the redundant Unnanmed: 0 column
df_2020.drop('Unnamed: 0' , axis=1 , inplace= True)
```

```python
# Converting cat data into numeric
df_2020_dummies = pd.get_dummies(df_2020)
df_2020_dummies.head()
```

```python
# Finding the column that is missing in our test data set

cols_2019 = list(df_2019_dummies.columns) # columns in 2019 dummy dataset

cols_2020 = list(df_2020_dummies.columns) # columns in 2020 dummy dataset

# Find the missing feature
missing_features = []

for feature in cols_2019:

    if feature not in cols_2020:
        missing_features.append(feature)
        
    
print(missing_features)
```

The missing data can be imputed from debt_settlement_flag_N due to the binary origin of the column
... That is, if debt_settlement_N is True then deb_settlement_Y is False ,
Knowing that we can create an artificial column of data so that our training and test sets have the same shape

```python
# Filling in the missing data in the debt_settlement_flag_Y column
new_col =  []
for value in df_2020_dummies['debt_settlement_flag_N']:

    if value == 1:
        new_col.append(0)
    else:
        new_col.append(1)

df_2020_dummies['debt_settlement_flag_Y'] = new_col
```

```python
X_test = df_2020_dummies.drop(['loan_status_low_risk' , 'loan_status_high_risk'] , axis=1)
y_test = df_2020_dummies[ 'loan_status_high_risk']
print(len(X_test))
print(len(y_test))
```

# Modeling
Random Forest:
* implicitly performs feature selection , making it great when there are a large number of features , we have 93 here
* bins variables making it resistant to outliers
* can handle linear and non linear relationships
* provide high accuracy and balance the bias-variance trade-off well
*  https://medium.datadriveninvestor.com/random-forest-pros-and-cons-c1c42fb64f04

Logistic Regression:
* weak against non-linear relationships but effective against linear ones
* sometimes simple is better and outputs are interpretable as probabilities
* https://www.geeksforgeeks.org/advantages-and-disadvantages-of-logistic-regression/

## Predictions
Since most data in the real world is not linear I have a hunch that vanilla logistic regression will underperform.


```python
# fitting the training data to our logistic and random forest classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Logistic Classfier
LR = LogisticRegression().fit(X_train , y_train)
print(f'Training Score: {LR.score(X_train , y_train)}')
print(f'Test Score: {LR.score(X_test , y_test)}')
```

```python
# Random Forest Classifier
RF = RandomForestClassifier().fit(X_train , y_train)
print(f'Training Score: {RF.score(X_train , y_train)}')
print(f'Test Score: {RF.score(X_test , y_test)}')
```

# Results
It looks like the random forest classifier performed better than Logistic Regression , but the perfect training score for the random forests result may indicate
Overfitting. Could it be that this dataset does not have a linear relationship like I predicted.

# Scaling

```python
from sklearn.preprocessing import StandardScaler
# Scaling the training data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Scaling Predicitions
Random Forest is an information based algorithm making immune to the effects of feature scaling , since logistic regression is an error based algorithm feature scaling will help it converge faster.

For that reason I believe that the score for our logistic regression model will improve while the score of our random forest classifier will stay the same
https://botbark.com/2021/01/26/why-random-forest-algorithm-is-not-affected-by-feature-scaling/ 

```python
# Fitting the models to the scaled data

# Random Forest
LRS = LogisticRegression().fit(X_train_scaled , y_train)
print(f'Score/ScaledScore: {LR.score(X_test , y_test)}/{LRS.score(X_test_scaled , y_test)}')

# Random Forest
RFS = RandomForestClassifier().fit(X_train_scaled , y_train)
print(f'Score/ScaledScore: {RF.score(X_test , y_test)}/{RFS.score(X_test_scaled , y_test)}')
```

# Scaling Results
As anticipated , the random forest algorithm stayed pretty much the same. However the logistic regression alogirthm got a big boost and even surpassed the random forest algorithm in accuracy. This shows that a properly tuned simple algorithm can outperform complicated algorithms in certain applications. In this case not only would the logistic regression model be more accurate but it would be less computer resource intensive.
